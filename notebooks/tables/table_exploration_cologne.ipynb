```python
levels = (0, 1, 2, 3, 8, 9, 10, 11)
contact_to_levels = {
    0 : '0R',
    1 : '1R',
    2 : '2R',
    3 : '3R',
    8 : '0L',
    9 : '1L',
    10 : '2L',
    11 : '3L'}


df = mapping[mapping['leadModel'] == 'Medtronic 3389']
df = df.copy()
df.loc[:, 'contactID'] = df['contactID'].map(contact_to_levels).astype(str)
df = df.rename(columns={'contactID' : 'level'})
df
```

```python
pivot_df = df.pivot_table(index='leadID', columns='level', aggfunc='size', fill_value=0)
pivot_df = pivot_df[['0L', '1L', '2L', '3L', '0R', '1R', '2R', '3R']]
sum = pivot_df.sum()[['0L', '1L', '2L', '3L', '0R', '1R', '2R', '3R']]
sum
```

```python
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import numpy as np

image = plt.imread('../../plots/external/Medronic3389Horizontal.png')

categories_L, categories_R = (['0L', '1L', '2L', '3L'], ['0R', '1R', '2R', '3R'])
values_L = sum[categories_L].to_numpy()
values_R = sum[categories_R].to_numpy()

fig, ax = plt.subplots(figsize=(8,4))

bar_width = 0.3
bar_positions_group1 = np.arange(len(categories_L))
bar_positions_group2 = bar_positions_group1 + bar_width

#image_extent = [0, len(categories_L), -0.4, -0.2]
ax.grid(axis = 'y', zorder = 0)

imagebox = OffsetImage(image, zoom=1.4)
ab = AnnotationBbox(imagebox, (1.7, 90), xycoords='data', frameon=False, pad = 0.3)
ax.add_artist(ab)

ax.bar(bar_positions_group1, values_L, color='black', width=bar_width, label='Left', zorder=3)
ax.bar(bar_positions_group2, values_R, color='black', width=bar_width, label='Right', zorder=3)

start, end = ax.get_ylim()
ax.yaxis.set_ticks(np.arange(start, end, 5))

group1_labels = ['Label 1', 'Label 2', 'Label 3']
group2_labels = ['Label A', 'Label B', 'Label C']
xtick_positions = np.concatenate([bar_positions_group1, bar_positions_group2])
xtick_labels = np.concatenate([categories_L, categories_R])

#ax.set_xlim(-0.5, len(categories_L+categories_R) - 0.5)
ax.set_ylim(0, max(values_L + values_R) + 30)
ax.set_xticks(xtick_positions, xtick_labels)
#ax.set_xlabel('Ring')
#ax.set_ylabel('Count')


plt.title('Cologne Medtronic 3389 Cohort levels mapping count', y= -0.2)
#plt.tight_layout()
plt.show()
```

```python
df_1.groups.keys()
```

```python
import matplotlib.pyplot as plt
import numpy as np

patients = [key[0] for key in df_1.groups.keys()]
contacts = [key[1] for key in df_1.groups.keys()]
max_contacts = np.max([np.unique(patients, return_counts=True)[1]])

unique_patients = np.unique(patients, return_counts=True)[0]
unique_contacts = np.unique(contacts, return_counts=True)[0]
```

```python
fig, axs = plt.subplots(figsize=(20,50), 
                        nrows=len(unique_patients), 
                        ncols=len(unique_contacts),   
                        gridspec_kw=dict(hspace=0.4))
keys = df_1.groups.keys()

for i, patient in enumerate(unique_patients):
    for j, contact in enumerate(unique_contacts):
        key = (patient, contact)
        if key in keys:
            axs[i,j].plot(
            df_1.get_group(key)['amplitude'],
            df_1.get_group(key)['clinicalScore'])

            axs[i,j].set_xticks(np.arange(0, 6, 1.0))
            axs[i,j].set_yticks(np.arange(0, 1.25, 0.25))
        else :
            axs[i,j].set_facecolor((0.6, 0.6, 0.6))
            axs[i,j].set_xticks([])
            axs[i,j].set_yticks([])
        title = ''
        if j == 0:
            title += f'Patient : {key[0]}'
        if i == 0:
            title += f'Contact : {key[1]}' if j != 0 else f', Contact : {key[1]}'
        axs[i,j].set_title(title, size=8)
        axs[i,j].set_ylim([0.0, 1.0])
        axs[i,j].set_xlim([0.0, 5.0])
        
ax.legend()
plt.show()
```
